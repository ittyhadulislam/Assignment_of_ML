```python
import pandas as pd
import seaborn as seaborn
```

```python
df = pd.read_csv('agora.csv')
```

```python
df.shape
```

```python
#Handle NaN values
mean = df.Transport.mean()
median = df.Transport.median()
```

```python
df['Transport'] = df['Transport'].fillna(median) #Fill NaN Values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df.Area = label.fit_transform(df['Area'])
```

```python
df.columns
```

```python
x = df.drop(['Profit'],axis=1)
```

```python
y = df['Profit']
```

```python
x.shape
```

# Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mmx = MinMaxScaler()
```

```python
newx = mmx.fit_transform(x)
```

```python
x = pd.DataFrame(newx, columns = ['Marketing Spend', 'Administration', 'Transport', 'Area'])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```
