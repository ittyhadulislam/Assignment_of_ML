```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("credit card.csv")
```

```python
data
```

```python
data = data.drop("ID", axis = 1)
```

```python
data
```

```python
x = data.drop("default.payment.next.month", axis = 1)
```

```python
y = data["default.payment.next.month"]
```

```python
x
```

# Feature Selection : ExtraTreeClassifier()

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
ex = ExtraTreesClassifier()
```

```python
ex.fit(x, y)
```

```python
ex.feature_importances_
```

```python
f = pd.Series(ex.feature_importances_)
```

```python
f
```

```python
f.nlargest(23)
```

```python
x
```

```python
corr = x.corr()
```

```python
ind = corr.index
```

```python
plt.figure(figsize = [20, 20])
sns.heatmap(x[ind].corr(), annot=True)
```

# Feature Selection : SelectKBest()

```python
from sklearn.feature_selection import SelectKBest, f_classif, f_regression
```

```python
skb = SelectKBest(score_func = f_classif)
```

```python
best_feature = skb.fit(x, y)
```

```python
best_feature.scores_
```

```python
df = pd.DataFrame(best_feature.scores_, columns=["Best Score"])
df1 = pd.DataFrame(x.columns, columns=["Columns"])
```

```python
new_data = pd.concat([df1, df], axis = 1)
```

```python
new_data
```

```python
new_data.nlargest(18, ["Best Score"])
```

# PCA

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
scaled_x = std.fit_transform(x)
```

```python
x
```

```python
scaled_x
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components= 3)
```

```python
x_pca = pca.fit_transform(scaled_x)
```

```python
x_pca
```

```python
df = pd.DataFrame(x_pca, columns= ["PCA - 1", "PCA - 2", "PCA - 3"])
```
